## –®–∞–≥ 3. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –º–æ–¥–µ–ª—å –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏—è –æ—Ç—Ç–æ–∫–∞ –∫–ª–∏–µ–Ω—Ç–æ–≤ <a class="anchor" id="third-bullet"></a>

<pre><b>–†–∞–∑–±–µ–π—Ç–µ –¥–∞–Ω–Ω—ã–µ –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –≤—ã–±–æ—Ä–∫—É —Ñ—É–Ω–∫—Ü–∏–µ–π train_test_split()</b></pre>

<div class="alert alert-warning">

**–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞**

–ú–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –ø–∞—Ä–∞–º–µ—Ç—Ä stratify = y –≤ —Ñ—É–Ω–∫—Ü–∏–∏ train_test_split( ), –≥–¥–µ y ‚Äì —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∞—è —Ü–µ–ª–µ–≤–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è. –î–∞–Ω–Ω—ã–µ —Ä–∞–∑–¥–µ–ª—è—Ç—Å—è –Ω–∞ –¥–≤–∞ –∫–ª–∞—Å—Ç–µ—Ä–∞ (–Ω–∞–±–ª—é–¥–µ–Ω–∏—è –∫–ª–∞—Å—Å–∞ 0 –∏ –Ω–∞–±–ª—é–¥–µ–Ω–∏—è –∫–ª–∞—Å—Å–∞ 1), –∏ –≤ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –≤—ã–±–æ—Ä–∫—É –ø–æ–ø–∞–¥—É—Ç –ø–æ 20% –Ω–∞–±–ª—é–¥–µ–Ω–∏–π –∏–∑ –∫–∞–∂–¥–æ–≥–æ –∫–ª–∞—Å—Ç–µ—Ä–∞:


    
–¢–æ –µ—Å—Ç—å, –µ—Å–ª–∏ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤ –≤ –æ—Ä–∏–≥–∏–Ω–∞–ª—å–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –±—ã–ª–æ 2:5, —Ç–æ –≤ –æ–±—É—á–∞—é—â–µ–π –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∞—Ö –æ–Ω–æ —Å–æ—Ö—Ä–∞–Ω–∏—Ç—Å—è —Ç–∞–∫–∏–º –∂–µ. –¢–∞–∫–æ–π –ø–æ–¥—Ö–æ–¥ –≤ —á–µ–º-—Ç–æ —Ä–µ—à–∞–µ—Ç –ø—Ä–æ–±–ª–µ–º—É –Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –∫–ª–∞—Å—Å–æ–≤.</div>

<div class="alert alert-info">

**–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞**

–ë–ª–∞–≥–æ–¥–∞—Ä—é, –ø–æ—á–∏—Ç–∞—é –ø—Ä–æ —ç—Ç–æ—Ç –º–æ–º–µ–Ω—Ç

```python
# —Ä–∞–∑–¥–µ–ª—è–µ–º –∏—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –Ω–∞ –ø—Ä–∏–∑–Ω–∞–∫–∏ (–º–∞—Ç—Ä–∏—Ü–∞ X) –∏ —Ü–µ–ª–µ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é (y)
X = df.drop('Churn', axis = 1)
y = df['Churn']

# —Ä–∞–∑–¥–µ–ª—è–µ–º –º–æ–¥–µ–ª—å –Ω–∞ –æ–±—É—á–∞—é—â—É—é (trian) –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é (test) –≤—ã–±–æ—Ä–∫—É
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# –æ–±—É—á–∞–µ–º scaler –∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∏—Ä—É–µ–º –º–∞—Ç—Ä–∏—Ü—É –¥–ª—è –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏
scaler = StandardScaler()
X_train_st = scaler.fit_transform(X_train)  
# –ø—Ä–∏–º–µ–Ω—è–µ–º —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏—é –∫ –º–∞—Ç—Ä–∏—Ü–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏
X_test_st = scaler.transform(X_test)
```

<pre><b>–û–±—É—á–∏—Ç–µ –º–æ–¥–µ–ª—å –Ω–∞ train-–≤—ã–±–æ—Ä–∫–µ –¥–≤—É–º—è —Å–ø–æ—Å–æ–±–∞–º–∏:
‚Äî –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–µ–π,
‚Äî —Å–ª—É—á–∞–π–Ω—ã–º –ª–µ—Å–æ–º.</b></pre>

<pre><b>–û—Ü–µ–Ω–∏—Ç–µ –º–µ—Ç—Ä–∏–∫–∏ accuracy, precision –∏ recall –¥–ª—è –æ–±–µ–∏—Ö –º–æ–¥–µ–ª–µ–π –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ. –°—Ä–∞–≤–Ω–∏—Ç–µ –ø–æ –Ω–∏–º –º–æ–¥–µ–ª–∏. –ö–∞–∫–∞—è –º–æ–¥–µ–ª—å –ø–æ–∫–∞–∑–∞–ª–∞ —Å–µ–±—è –ª—É—á—à–µ –Ω–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –º–µ—Ç—Ä–∏–∫?</b></pre>

### –õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è

```python
# –∑–∞–¥–∞—ë–º –∞–ª–≥–æ—Ä–∏—Ç–º –¥–ª—è –Ω–∞—à–µ–π –º–æ–¥–µ–ª–∏

model_lr = LogisticRegression(solver='lbfgs', random_state=0) # –∑–∞–¥–∞–π—Ç–µ –∞–ª–≥–æ—Ä–∏—Ç–º –¥–ª—è –Ω–∞—à–µ–π –º–æ–¥–µ–ª–∏
```

```python
# –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å
model_lr.fit(X_train_st, y_train)
```

```python
# –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è —É–∂–µ –æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª—å—é, —á—Ç–æ–±—ã —Å–¥–µ–ª–∞—Ç—å –ø—Ä–æ–≥–Ω–æ–∑—ã
predictions_lr = model_lr.predict(X_test_st) 
```

```python
# –≤—ã–≤–µ–¥–µ–º –æ—Å–Ω–æ–≤–Ω—ã–µ –º–µ—Ç—Ä–∏–∫–∏
print('Accuracy: {:.2f}'.format(accuracy_score(y_test,predictions_lr)))
print('Precision: {:.2f}'.format(precision_score(y_test,predictions_lr)))
print('Recall: {:.2f}'.format(recall_score(y_test,predictions_lr)))
```

### –°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å

```python
# –∑–∞–¥–∞—ë–º –∞–ª–≥–æ—Ä–∏—Ç–º –¥–ª—è –Ω–∞—à–µ–π –º–æ–¥–µ–ª–∏

rf_model = RandomForestClassifier(random_state = 0)
```

```python
# –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞
rf_model.fit(X_train_st, y_train)
```

```python
# –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è —É–∂–µ –æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª—å—é, —á—Ç–æ–±—ã —Å–¥–µ–ª–∞—Ç—å –ø—Ä–æ–≥–Ω–æ–∑—ã
rf_predictions = rf_model.predict(X_test_st)
```

```python
# –≤—ã–≤–µ–¥–µ–º –æ—Å–Ω–æ–≤–Ω—ã–µ –º–µ—Ç—Ä–∏–∫–∏
print('Accuracy: {:.2f}'.format(accuracy_score(y_test,rf_predictions)))
print('Precision: {:.2f}'.format(precision_score(y_test,rf_predictions)))
print('Recall: {:.2f}'.format(recall_score(y_test,rf_predictions)))
```

```python
# –∑–∞–¥–∞—ë–º –º–æ–¥–µ–ª—å k_means —Å —á–∏—Å–ª–æ–º –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ 5 –Ω–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –¥–µ–Ω–¥—Ä–æ–≥—Ä–∞–º–º—ã –≤—ã—à–µ
kmeans = KMeans(n_clusters = 5, random_state=0)
# –ø—Ä–∏–º–µ–Ω—è–µ–º –∞–ª–≥–æ—Ä–∏—Ç–º –∫ –¥–∞–Ω–Ω—ã–º –∏ —Ñ–æ—Ä–º–∏—Ä—É–µ–º –≤–µ–∫—Ç–æ—Ä –∫–ª–∞—Å—Ç–µ—Ä–æ–≤
# –í –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π labels —Å–æ—Ö—Ä–∞–Ω—è—é—Ç—Å—è –∏–Ω–¥–µ–∫—Å—ã –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–Ω—ã—Ö –∞–ª–≥–æ—Ä–∏—Ç–º–æ–º –≥—Ä—É–ø–ø
labels = kmeans.fit_predict(x_sc)
```

</div>
